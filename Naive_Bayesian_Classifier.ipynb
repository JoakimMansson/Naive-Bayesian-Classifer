```python
from sklearn.datasets import load_iris
from copy import deepcopy
iris = load_iris()
X, y = iris["data"], iris["target"]

#training_X, training_y = X[:int(len(X)*0.7)], y[:int(len(X)*0.7)] # 70% training data
# Index 0: Sepal length, Index 1: Sepal width, Index 2: Petal length, Index 3: Petal width

print("data", X)
print("class/label", y)
```

```python
flower_1 = {"sepal length (cm)": {}, "sepal width (cm)": {}, "petal length (cm)": {}, "petal width (cm)": {}}
flower_2 = {"sepal length (cm)": {}, "sepal width (cm)": {}, "petal length (cm)": {}, "petal width (cm)": {}}
flower_3 = {"sepal length (cm)": {}, "sepal width (cm)": {}, "petal length (cm)": {}, "petal width (cm)": {}}
flowers = {"flower_1": flower_1, "flower_2": flower_2, "flower_3": flower_3}

# Fill flowers with a count of each occuring value for each feature
for r, row in enumerate(X):
    for c, col in enumerate(row):
        if y[r] == 0:
            if col not in flowers["flower_1"][iris["feature_names"][c]]:
                flowers["flower_1"][iris["feature_names"][c]][col] = 1
            else:
                flowers["flower_1"][iris["feature_names"][c]][col] += 1
        elif y[r] == 1:
            if col not in flowers["flower_2"][iris["feature_names"][c]]:
                flowers["flower_2"][iris["feature_names"][c]][col] = 1
            else:
                flowers["flower_2"][iris["feature_names"][c]][col] += 1
        elif y[r] == 2:
            if col not in flowers["flower_3"][iris["feature_names"][c]]:
                flowers["flower_3"][iris["feature_names"][c]][col] = 1
            else:
                flowers["flower_3"][iris["feature_names"][c]][col] += 1

print("flowers", flowers)

# Fill flowers with a probability of each occuring value for each feature
for flower in flowers:
    for feature in flowers[flower]:
        for value in flowers[flower][feature]:
            flowers[flower][feature][value] = flowers[flower][feature][value]/50

print("flowers with probability for each occurence: ", flowers)
# ++ Test sepal_length summed up probability ++ #
print("Probability of sepal length 5.1 for flower 1 SUM: ", sum(flowers["flower_1"]["sepal length (cm)"][key] for key in flowers["flower_1"]["sepal length (cm)"]))
# -- Test sepal_length summed up probability -- #
```

```python
def predict_flower(sepal_length, sepal_width, petal_length, petal_width):
    flower_1_probability = 1
    flower_2_probability = 1
    flower_3_probability = 1
    for f, flower in enumerate(flowers.keys()):
        for feature in flowers[flower]:
            if feature == "sepal length (cm)":
                if f == 0:
                    if sepal_length in flowers[flower][feature]:
                        flower_1_probability *= flowers[flower][feature][sepal_length]
                    elif flower_1_probability > 0:
                        flower_1_probability *= -1
                        break
                elif f == 1:
                    if sepal_length in flowers[flower][feature]:
                        flower_2_probability *= flowers[flower][feature][sepal_length]
                    elif flower_2_probability > 0:
                        flower_2_probability *= -1
                        break
                else:
                    if sepal_length in flowers[flower][feature]:
                        flower_3_probability *= flowers[flower][feature][sepal_length]
                    elif flower_3_probability > 0:
                        flower_3_probability *= -1
                        break
            elif feature == "sepal width (cm)":
                if f == 0:
                    if sepal_width in flowers[flower][feature]:
                        flower_1_probability *= flowers[flower][feature][sepal_width]
                    elif flower_1_probability > 0:
                        flower_1_probability *= -1
                        break
                elif f == 1:
                    if sepal_width in flowers[flower][feature]:
                        flower_2_probability *= flowers[flower][feature][sepal_width]
                    elif flower_2_probability > 0:
                        flower_2_probability *= -1
                        break
                else:
                    if sepal_width in flowers[flower][feature]:
                        flower_3_probability *= flowers[flower][feature][sepal_width]
                    elif flower_3_probability > 0:
                        flower_3_probability *= -1
                        break
            elif feature == "petal length (cm)":
                if f == 0:
                    if petal_length in flowers[flower][feature]:
                        flower_1_probability *= flowers[flower][feature][petal_length]
                    elif flower_1_probability > 0:
                        flower_1_probability *= -1
                        break
                elif f == 1:
                    if petal_length in flowers[flower][feature]:
                        flower_2_probability *= flowers[flower][feature][petal_length]
                    elif flower_2_probability > 0:
                        flower_2_probability *= -1
                        break
                else:
                    if petal_length in flowers[flower][feature]:
                        flower_3_probability *= flowers[flower][feature][petal_length]
                    elif flower_3_probability > 0:
                        flower_3_probability *= -1
                        break
            elif feature == "petal width (cm)":
                if f == 0:
                    if petal_width in flowers[flower][feature]:
                        flower_1_probability *= flowers[flower][feature][petal_width]
                    elif flower_1_probability > 0:
                        flower_1_probability *= -1
                        break
                elif f == 1:   
                    if petal_width in flowers[flower][feature]:
                        flower_2_probability *= flowers[flower][feature][petal_width]
                    elif flower_2_probability > 0:
                        flower_2_probability *= -1
                        break 
                else:
                    if petal_width in flowers[flower][feature]:
                        flower_3_probability *= flowers[flower][feature][petal_width]
                    elif flower_3_probability > 0:
                        flower_3_probability *= -1
                        break
    print("flower_1_probability: ", flower_1_probability)
    print("flower_2_probability: ", flower_2_probability)
    print("flower_3_probability: ", flower_3_probability)
    if flower_1_probability < 0 and flower_2_probability < 0 and flower_3_probability < 0:
        return None
    elif flower_1_probability > flower_2_probability and flower_1_probability > flower_3_probability:
        return 0
    elif flower_2_probability > flower_1_probability and flower_2_probability > flower_3_probability:
        return 1
    elif flower_3_probability > flower_1_probability and flower_3_probability > flower_2_probability:
        return 2


# If None is returned, then some flower details have not been seen before
print("Predicted flower species ", predict_flower(sepal_length=5.1, sepal_width=3.5, petal_length=1.4, petal_width=0.2)) # Should be 0
print("------------------------------------------")
print("Predicted flower species ", predict_flower(sepal_length=7.2, sepal_width=3.0, petal_length=5.8, petal_width=1.6)) # Should be 2
print("------------------------------------------")
print("Predicted flower species ", predict_flower(sepal_length=5.1, sepal_width=2.5, petal_length=3.0, petal_width=1.1)) # Should be 1
print("------------------------------------------")
print("Predicted flower species ", predict_flower(sepal_length=5.0, sepal_width=3.0, petal_length=1.6, petal_width=0.2)) # Should be 0
print("------------------------------------------")
print("Predicted flower species ", predict_flower(sepal_length=5.9, sepal_width=3.0, petal_length=5.1, petal_width=1.8)) # Should be 2
```
